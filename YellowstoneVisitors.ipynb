```
import pandas as pd
```

```
# yellowstone visitors
# data from https://data.world/nps/yellowstone-np
# source: https://irma.nps.gov/Stats/

visitors = pd.read_csv('https://query.data.world/s/nkqhdbl6wfdt2yws7dij5fpueptmo4', thousands=',')

visitors.rename(columns={'Total':'Total Visitors'}, inplace=True)

visitors.head()
```

```
visitors.isnull().sum()
```

```
# global population from https://www.worldometers.info/world-population/world-population-by-year/
# US Population data from https://www.multpl.com/united-states-population/table/by-year and estimated using census data
# CSV created by inputting these data into a google sheet and coverted into an excel csv


population = pd.read_csv('https://raw.githubusercontent.com/JraHike76/Build-Week-Unit-1/master/Build%20Week%20Files/world_US_pop_1979-2018%20-%20Sheet1.csv')
population.head()
```

```
population.isnull().sum()
```

```
# merge dataframes to add world population and US population by year to visitor dataframe

visitors_pop = pd.merge(visitors,population)

visitors_pop 
```

```
#getting additional information using the describe function

visitors_pop.info()
```

```
# Adding a column to show the percentage of visitors from the world
# Adding a column to show the percentage of visitors from the US

visitors_pop['Percent of World Visitors'] = (100*(visitors_pop['Total Visitors'])/visitors_pop['World Population']).round(2)
visitors_pop['Percent of US Visitors'] = (100*(visitors_pop['Total Visitors'])/visitors_pop['US Population']).round(2)
visitors_pop
```

```
visitors_pop.info()
```

```
# I created a new csv with the merged dataframe to make it easier to work with

visitors_pop.to_csv('visitor.csv')
```

```
# Explanatory Anaysis
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

visitor_pop = pd.read_csv('visitor.csv')
visitor_pop
```

```
# Explanatory Anaysis

# To create a better data telling story I have to re-order the rows by years ascending 

visitors_sort = visitor_pop.sort_values(by ='Year' , ascending=True)
 
visitors_sort
```

```
# now I will drop the new column that was created titled 'Unnamed :0' that contains the out of order indeces

visitors_y = visitors_sort.drop(['Unnamed: 0'], axis=1)
visitors_y
```

```
# Sorting values of Total Visitors for plotting
visitors_y['Total Visitors'].sort_values()
```

```
# importing libraries 
import matplotlib.pyplot as plt 
import numpy as np

# Simple Line graph showing total number of visitors 1979 - 2018

# Plot size
fig, ax = plt.subplots(figsize=(10,5))

# values of x and y axes 
x = np.array(visitors_y['Year']) 
y = np.array(visitors_y['Total Visitors']) 
  
plt.plot(x, y, color='green')
# x and y labels 
plt.xlabel('Year', fontsize=14, fontweight='bold') 
plt.ylabel('Visitors (billions)', fontsize=14, fontweight='bold') 
  
# Set x and y ticks 
plt.xticks(np.arange(1978, 2020, 10), fontsize=11)
plt.yticks(np.arange(1800000, 4800000, 500000), fontsize=11)  

# Set Title
plt.title('Yellowstone Visitors 1979 - 2018', fontsize=18, fontweight='bold', x=.48, y=1.12)

# Add footer
txt='source: https://irma.nps.gov/Stats/'
plt.figtext(0.2, -0.03, txt, wrap=True, horizontalalignment='center', fontsize=8)

#remove unnecessary text
#plt.remove(findobj(text = 'le6'))
#I guess Paint will do the job

plt.show() 
```

```
# Now I am going to create a bar plot showing vistors in January(least) vs visitors in July(most)

# I only want certain years of data to represent 1979, 1989, 1999, 2009, 2018
# So I will create a new dataframe called vis_c to represent the condensed dataframe
# I will keep only rows necessary for this bar plot

vis_c = visitors_y.loc[[0, 9, 19, 29, 39]].sort_values(by ='Year' , ascending=True)
vis_c
```

```
# Checking July for my high value for my y axis

vis_c['JUL'].sort_values()
```

```
# my range of values for the y axis will be 0 - 950,000
# which will include the highest value in July
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# renaming dataframe df for efficiency
df = vis_c

# Setting the positions and width for the bars
pos = list(range(len(df['Year']))) 
width = 0.4 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with JAN in position pos,
plt.bar(pos, df['JAN'], width, alpha=0.8, 
        label=df['Year'][0:39]) 

# Create a bar with JUL in position pos,
plt.bar([p + width for p in pos], df['JUL'], width, alpha=0.8, 
        label=df['Year'][0:39]) 

# Set the axis labels
ax.set_ylabel('Visitors', fontsize = 14, fontweight = 'bold')
ax.set_xlabel('Year', fontsize = 14, fontweight = 'bold')

# Set the chart's title
ax.set_title('Yellowstone Visitors Winter vs. Summer', fontsize = 16, fontweight = 'bold')

# Set the position of the x ticks
ax.set_xticks([p + 0.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['Year'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(df['JAN'] + df['JUL'])])

# Adding the legend
plt.legend(['January', 'July'],fontsize = 14, loc='upper left')

# Add footer
txt='source: https://irma.nps.gov/Stats/'
plt.figtext(0.15, 0, txt, wrap=True, horizontalalignment='center', fontsize=8)

# Showing grid
plt.grid()


plt.show()
```

```
# During the least busiest month, January there does not seem to be any signicance to increase or decrease of visitors 
# During the busiest month, July there is a signicant increase of visitors

# Reasons for stagnation in winter?:
# Winter access is limited to snowshoes/x-country skiing
# Extreme cold temperatures, high winds and frequent snowstorms

# Well duh, who wants to enjoy nature in these conditions? Or maybe they are all enjoying the warmer weather in the Southern Hemisphere
```

```
# Let's not leave out the Spring and Fall season!
# Now I am going to create a bar plot showing vistors in April vs visitors in October

# This will give me a better sense of what the first nice month after winter looks like
# Versus the first month it starts to get cold enough to snow  

# Much like January and July represented Winter and Summer
# May and September will represent Spring and Fall
# Since Spring comes late and Fall comes early at that elevation

# My theory is that these months will have more in common than January and July
# And both show an increase in visitors over the years
```

```
# I need to find my high value for my Y axis, this time I will have to check both months

# May's values

vis_c['MAY'].sort_values()
```

```
# September's values

vis_c['SEP'].sort_values()
```

```
# September has higher values! With 724,454 as the highest. My range for y will be 0 - 750,000
```

```
# Bar plot showing vistors in May vs visitors in September

# Will still use dataframe df for efficiency
df = vis_c

# Setting the positions and width for the bars
pos = list(range(len(df['Year']))) 
width = 0.4 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with MAY in position pos,
plt.bar(pos, df['MAY'], width, alpha=0.8, 
        label=df['Year'][0:39]) 

# Create a bar with SEP in position pos,
plt.bar([p + width for p in pos], df['SEP'], width, alpha=0.8, 
        label=df['Year'][0:39]) 

# Set the axis labels
ax.set_ylabel('Visitors', fontsize = 14, fontweight = 'bold')
ax.set_xlabel('Year', fontsize = 14, fontweight = 'bold')

# Set the chart's title
ax.set_title('Yellowstone Visitors: Spring vs. Fall', fontsize = 16, fontweight = 'bold')

# Set the position of the x ticks
ax.set_xticks([p + 0.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['Year'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
# For uniformity I am using the same range for y as above graph
plt.ylim([0, max(df['JAN'] + df['JUL'])])

# Adding the legend
plt.legend(['May', 'September'],fontsize = 14, loc='upper left')

# Add footer
txt='source: https://irma.nps.gov/Stats/'
plt.figtext(0.15, 0, txt, wrap=True, horizontalalignment='center', fontsize=8)

# Showing grid
plt.grid()


plt.show()
```

```
# The results represented in the graph show a significant increase in visitors for both months
# However, May had a much lower visitor count than September

# Possible reasons for the difference:
# May is most likely a very wet month leaving camgrounds and trails a muddy mess
# September might bring out the leaf peepers and those taking last minute family vacations before school starts
# Also though it is cooler September is probably the ideal month to visit given there is significantly less than 4 million visitors 
```

```
# Line graph showing the percent of US visitors and percent of world visitors

# Is there an increase in interest of visiting yellowstone??

import matplotlib.pyplot as plt
import pandas as pd

# Plot size
fig, ax = plt.subplots(figsize=(10,5))

# set ax to get current axis on the plot
ax = plt.gca()

# Set the axis labels
ax.set_ylabel('Visitors (%)', fontsize = 14, fontweight = 'bold')
ax.set_xlabel('Year', fontsize = 14, fontweight = 'bold')

# Set the chart's title
ax.set_title('Increase in Visitors: US vs. World', fontsize = 16, fontweight = 'bold')

visitors_y.plot(kind='line',x='Year',y='Percent of World Visitors',ax=ax)
visitors_y.plot(kind='line',x='Year',y='Percent of US Visitors',ax=ax)

# Add footer
txt='source: https://www.worldometers.info/world-population/world-population-by-year/ and https://www.multpl.com/united-states-population/table/by-year'
plt.figtext(0.5, -.05, txt, wrap=True, horizontalalignment='center', fontsize=8)

plt.show()
```

```
# There is a slight increase in the percentage of the US population but only about 0.5%
# The percentage of the world that visited Yellowstone showed no increase 
# Conclusion: The world may be increasing with a love for nature but not necessarily here at Yellowstone
# and the slight increase in US interest in Yellowstone may be due to it's proximity
# Oh and carrying your poop down Mt. Rainier, it's true, Google it.
```
